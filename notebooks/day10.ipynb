```python
points_map = {
    ')': 3,
    ']': 57,
    '}': 1197,
    '>': 25137,
}

brackets = {
    '(': ')',
    '{': '}',
    '[': ']',
    '<': '>',
}

rev_brackets = {v: k for k, v in brackets.items()}
```

```python
def find_illegal(s):
    stack = []
    for c in s:
        if c in brackets:
            stack.append(c)
        elif stack and stack[-1] == rev_brackets[c]:
            stack.pop()
        else:
            return c
```

```python
from collections import Counter

counts = Counter(find_illegal(i) for i in inp)
sum(points_map.get(k, 0) * v for k, v in counts.items())
```

```python
completion_points = {
    ')': 1,
    ']': 2,
    '}': 3,
    '>': 4,
}
```

```python
def find_incomplete(s):
    stack = []
    for c in s:
        if c in brackets:
            stack.append(c)
        elif stack and stack[-1] == rev_brackets[c]:
            stack.pop()
        else:
            return []
    return [brackets[i] for i in reversed(stack)]

def score_incomplete(lst):
    score = 0
    for i in lst:
        score *= 5
        score += completion_points[i]
    return score
```

```python
scores = [score_incomplete(lst) for expr in inp for lst in [find_incomplete(expr)] if lst]
scores.sort()
scores[len(scores) // 2]
```
