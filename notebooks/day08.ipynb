````python
def decode(line):
    digits, output = line
    digits = [tuple(sorted(set(i))) for i in digits]
    output = [tuple(sorted(set(i))) for i in output]
    
    decoded = {}
    known = {}
    def mark(code, meaning):
        decoded[code] = meaning
        known[meaning] = code
    
    # 1, 4, 7, 8 have known number of segments
    for i in digits:
        if len(i) == 2:
            mark(i, 1)
        elif len(i) == 4:
            mark(i, 4)
        elif len(i) == 3:
            mark(i, 7)
        elif len(i) == 7:
            mark(i, 8)
    
    # find 3, 6, 9
    for i in digits:
        # 3 has 5 segments and all of 1's segments
        if len(i) == 5 and set(i).issuperset(set(known[1])):
            mark(i, 3)
        # 6 and 9 have 6 segments
        elif len(i) == 6:
            # 6 doesn't have all of 1's segments
            if not set(i).issuperset(set(known[1])):
                mark(i, 6)
            # 9 has all of 4's segments
            elif set(i).issuperset(set(known[4])):
                mark(i, 9)
    
    # 0 is the last one with 6 segments
    for i in digits:
        if len(i) == 6 and i not in decoded:
            mark(i, 0)
    
    # 5 sectors intersected with 6 only leaves 5's sectors
    for i in digits:
        if len(i) == 5 and i not in decoded and set(i).intersection(set(known[6])) == set(i):
            mark(i, 5)
    
    # find 2 the last not decoded one
    for i in digits:
        if i not in decoded:
            mark(i, 2)
    
    val = 0
    for i in output:
        val *= 10
        val += decoded[i]
    return val
````
