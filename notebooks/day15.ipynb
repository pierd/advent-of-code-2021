```rust
use std::collections::VecDeque;

const MOVES: &[(isize, isize)] = &[
    (1, 0),
    (-1, 0),
    (0, 1),
    (0, -1),
];

fn solve(map: &Vec<Vec<usize>>) -> usize {
    let mut cost: Vec<Vec<usize>> = Vec::with_capacity(map.len());
    for _ in 0..map.len() {
        cost.push(vec![usize::MAX; map[0].len()]);
    }
    cost[0][0] = 0;
    let mut queue = VecDeque::new();
    queue.push_back((0, 0));
    while let Some((row, col)) = queue.pop_front() {
        for (drow, dcol) in MOVES {
            let target_row = row as isize + drow;
            let target_col = col as isize + dcol;
            if 0 <= target_row && 0 <= target_col {
                let target_row = target_row as usize;
                let target_col = target_col as usize;
                if target_row < cost.len() && target_col < cost[0].len() {
                    let target_cost = cost[row][col] + map[target_row][target_col];
                    if target_cost < cost[target_row][target_col] {
                        cost[target_row][target_col] = target_cost;
                        queue.push_back((target_row, target_col));
                    }
                }
            }
        }
    }
    *cost.last().unwrap().last().unwrap()
}
```

```rust
fn next_field(x: usize) -> usize {
    if x == 9 {
        1
    } else {
        x + 1
    }
}

fn extend(map: &Vec<Vec<usize>>) -> Vec<Vec<usize>> {
    let mut result = Vec::with_capacity(map.len() * 5);
    for mul in 0..5 {
        for row in map {
            let mut new_row = Vec::with_capacity(row.len() * 5);
            for i in row {
                let mut x = *i;
                for _ in 0..mul {
                    x = next_field(x);
                }
                new_row.push(x);
            }
            for i in row.len()..(row.len() * 5) {
                new_row.push(next_field(new_row[i - row.len()]));
            }
            result.push(new_row);
        }
    }
    result
}
```
